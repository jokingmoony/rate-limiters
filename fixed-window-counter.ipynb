```python
from threading import Thread, Lock
from datetime import datetime
import time

COUNTER_PER_SEC = 1
REQUEST_PER_SEC = 30
WINDOW_SIZE_SEC = 10

time_window = None
counter = 0

lock = Lock()

def reset_counter():
    global counter, time_window
    if time_window == f'{datetime.now():%H%M%S}':
        return
    with lock:
        counter = 0
        time_window = f'{datetime.now():%H%M%S}'

def process():
    global counter
    reset_counter()
    with lock:
        if counter >= WINDOW_SIZE_SEC:
            return False
        counter += 1
        return True

def request_in_sec():
    while True:
        time.sleep(1 / REQUEST_PER_SEC)
        res = process()
        if res:
            print(f'{datetime.now():%H%M%S}/ success')

            
request_thread = Thread(target=request_in_sec)
request_thread.start()
```
