```python
from datetime import datetime, timedelta
from threading import Thread, Lock
import time

log_stack = []

LIMIT = 10
lock = Lock()
REQUESTS_PER_SEC = 20

def process():
    global log_stack
    with lock:
        time_window_front = datetime.now() - timedelta(seconds = 1)
        valid_count = len(list(filter(lambda x : x >= time_window_front, log_stack)))
        # Remove Unused logs
        log_stack = log_stack[len(log_stack) - valid_count:]
        if valid_count < LIMIT:
            log_stack.append(datetime.now())
            return True
        return False

def request():
    res = process()
    if res:
        print(f'{datetime.now():%H%M%S}/ request success')

def request_in_sec():
    while True:
        time.sleep(1 / REQUESTS_PER_SEC)
        request()

request_thread = Thread(target=request_in_sec)
request_thread.start()







```
